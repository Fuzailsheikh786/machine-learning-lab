```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
print("Before adding missing values:\n", df.isnull().sum())
```

```python
df.head(10)
```

```python
df.iloc[0, 0] = np.nan
df.iloc[1, 2] = np.nan
df.iloc[2, 4] = np.nan
```

```python
print("\nAfter adding missing values:\n", df.isnull().sum())
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df_mean = df.fillna(df.mean(numeric_only=True))
print("\nData after filling with MEAN:\n")
print(df_mean.head(10))
```

```python
df_median = df.fillna(df.median(numeric_only=True))
print("\nData after filling with MEDIAN:\n")
print(df_median.head(10))
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['HouseAge'], color='salmon')
plt.title('Box Plot of House Age')
plt.ylabel('House Age (years)')
plt.grid(True)
plt.show()
```
